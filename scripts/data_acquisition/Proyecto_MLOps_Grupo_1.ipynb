A continuación, se muestran algunos ejemplos de 3 categorías de flores diferentes para que se pueda apreciar el tipo de imágenes que serán utilizados en el proyecto

```
#@title Ejemplos de la categoria 83 (Hibiscus)
#print("Ejemplos imagenes de la categoria 83 \n")

dir_path = "dataset/102 flower/flowers/train"
rutas_imagenes = [
    dir_path+"/83/"+i for i in [
        "image_01696.jpg", "image_01702.jpg",
        "image_01698.jpg", "image_01699.jpg"
    ]
]
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

for i, ruta_imagen in enumerate(rutas_imagenes):
    # Carga la imagen
    imagen = mpimg.imread(ruta_imagen)
    axs[i].imshow(imagen)
    axs[i].axis('off')
```
